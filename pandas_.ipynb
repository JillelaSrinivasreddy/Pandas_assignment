```python
import pandas as pd
```

```python
df = pd.read_excel("/Movies_Dataset.xlsx")
```

##1.Select all movies in the "action" genre


```python

df[df['Genre'] == 'Action']
```

##Select movies released after 2000


```python
df[df['Release_Year']> 2000]
```

## Select movies with an IMBD rating grater than 8.5


```python
df[df['IMDb_Rating'] > 8.5]
```

## Select only "Movie Name and "Genre" columns.


```python
df[['Movie Name', 'Genre']]
```

## Select movies released betwwen 1990 and 2010


```python
df[(df['Release_Year'] >= 1900)&(df['Release_Year'] <= 2010)]
```

## Count the number of movies in each genre

```python
df['Genre'].value_counts()
```

##Find the highest IMDb-rated movie.

```python
df[df['IMDb_Rating'] == df['IMDb_Rating'].max()]
```

## Find the average IMDb rating of all movies.



```python
df['IMDb_Rating'].mean()
```

##Find the total Revenue of all movies

```python
df['Revenue'].sum()
```

## Select movies with Revenue greater than 500 million


```python
df[df['Revenue'] > 500]
```

## Rank movies based on IMDb rating

```python
df['Rank'] = df['IMDb_Rating'].rank(ascending=False)
df.sort_values('Rank')
```

## Sort movies by release year in descending order

```python
df.sort_values('Release_Year', ascending = False)
```

##  Find all duplicate movie names (if any )

```python
df[df.duplicated(subset=['Movie Name'])]
```

## Remove  duplicate movie names

```python
df = df.drop_duplicates(subset=['Movie Name'])
```

## Replace "Sci-Fi" genre with "Science Fiction".

```python
df['Genre'] = df['Genre'].replace('Sci-Fi', 'Science Fiction')
```
